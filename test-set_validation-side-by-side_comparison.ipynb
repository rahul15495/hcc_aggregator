```python
def get_io(id):
    temp_patient= dict(df_person.loc[id])
    temp_diag= df_diag.query("Patient_ID=='{}'".format(temp_patient['Patient_ID']))

    lob= 'ACO'

    calc = api.single_month_score_predcitor(hicno=temp_patient['Patient_ID'],
                                 sex=temp_patient['Gender'],
                                 dob=temp_patient['Patient_DOB'],
                                 month_of_eligibility= '09',
                                 year_of_eligibility= '2018',
                                 RAF_type= temp_patient['RAF_Type'],
                                 lob= lob,
                                 orec=0,
                                 medicaid=True,
                                 codes=list(temp_diag[' ICD10'].values))
    
    sas= []
    for obs in list(set(df_output.query("Patient_ID=='{}'".format(temp_patient['Patient_ID']))['Observation'])):
        a=set(df_output.query("Patient_ID=='{}' and Observation=='{}'".format(temp_patient['Patient_ID'], obs))['RAF coefficient'])
        sas.append({obs: a.pop()})
        
    return calc, sas, temp_patient
```
